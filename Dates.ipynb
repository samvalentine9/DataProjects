```python
import csv
from datetime import datetime, timedelta
import math

# Set the start and end dates, create weekdays list
start_date = datetime(2009, 12, 1)
end_date = datetime(2011, 12, 31)
weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Create an empty list to store the data
dates_data = []

# Initialize the period counter
period = 0
date_id = 1

# Iterate through the dates and populate the list
current_date = start_date
while current_date <= end_date:
    date_str = current_date.strftime("%Y-%m-%d")
    year = current_date.year
    month = current_date.month
    day = current_date.day
    day_of_week_num = current_date.weekday() + 1 #this works cool as 0 is monday so it's already ready for the list index
    #, but we want to use the integer in the final script so we have to adjust
    day_of_week = weekdays[day_of_week_num-1] 
    quarter = math.ceil(month / 3)
    half = math.ceil(month / 6)
    period_str = period
    week_num = current_date.isocalendar()[1]  # Get the week number
    # Increment the period counter if it's the start of a new month
    if current_date.day == 1:
        period += 1
    dates_data.append([date_str, date_id, period, year, quarter, half, month, week_num, day, day_of_week, day_of_week_num])


        

    current_date += timedelta(days=1)
    date_id += 1

```

```python
print(dates_data)
```

```python
# Open the CSV file for writing
with open("Dates.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)

    # Write the header row
    writer.writerow(["Date", "Date ID", "Period", "Year", "Quarter", "Half", "Month", "Week", "Day", "Day of the Week", "Day of the Week n"])

    # Write the data rows
    for row in dates_data:
        writer.writerow(row)

print("Data exported to Dates.csv successfully.")
```
